# Question 1
## A1.Load the recorded speech file into your python workspace. Once loaded, plot the graph for the speech signal.You may use the below code from librosa asa reference.

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
# load an opus vorbis file
filename =r"C:\Users\THADAKALURU JASWANTH\Downloads\AISpeechProcessing\AIvoice_lab1.opus"
##load the speech signal
y, sr = librosa.load(filename)
print(y)
print(sr)
plt.figure(figsize=(16, 6))
librosa.display.waveshow(y)
plt.title('Speech Signal')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.show()
```
