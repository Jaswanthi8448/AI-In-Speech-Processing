```python
import librosa
import sounddevice as sd

# Load the audio file
y, sr = librosa.load(r"C:\Users\THADAKALURU JASWANTH\Downloads\Sem 6 LAB\AISpeechProcessing\LAB_03Voice.opus")

# Split the speech signal into segments based on detected silences
segments = librosa.effects.split(y, top_db=20)  # Adjust top_db as needed

# Listen to each segment
for i, segment in enumerate(segments):
    start, end = segment
    segment_audio = y[start:end]

    print(f"Playing Segment {i+1}: Duration {len(segment_audio)/sr} seconds")
    sd.play(segment_audio, sr)
    sd.wait()
```
